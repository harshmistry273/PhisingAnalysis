````python
corr = df.corr(numeric_only=True)
corr

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
sns.heatmap(corr, annot=False, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
````
