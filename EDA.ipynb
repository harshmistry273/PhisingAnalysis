### üî† This query plots the average count of special characters in URLs, comparing phishing vs legitimate websites.  


````python
special_chars_analysis = [
    "nb_dots", "nb_hyphens", "nb_at", "nb_qm", "nb_and", "nb_or",
    "nb_eq", "nb_underscore", "nb_tilde", "nb_percent", "nb_slash",
    "nb_star", "nb_colon", "nb_comma", "nb_semicolumn", "nb_dollar",
    "nb_space", "nb_www", "nb_com", "nb_dslash"
]

mean_values = df.groupby("status")[special_chars_analysis].mean().T

mean_values.plot(kind="bar", figsize=(14,6))
plt.grid(True)
plt.title("Average Special Characters Count in URLs by Class")
plt.ylabel("Mean Count")
plt.xticks(rotation=45)
````

The code calculates the average count of various special characters in URLs for legitimate vs phishing sites and visualizes the comparison using a bar plot.

Phishing and legitimate websites were categorized by traffic using web_traffic ranks: High, Medium, Low, and No Traffic. Pie charts show that most phishing sites fall in the low traffic category, while legitimate sites are mostly high or medium traffic. This indicates phishing domains are generally less popular, helping identify potentially risky sites.


To get the Categorical values we have created categories using the following code:

### üö¶ This query categorizes websites into High, Medium, or Low traffic based on their web traffic rank.  


````python
def categorize_traffic(rank):
    if rank <= 100000:
        return "High Traffic"
    elif rank <= 1000000:
        return "Medium Traffic"
    else:
        return "Low Traffic"

df["traffic_category"] = df["web_traffic"].apply(categorize_traffic)
````

### ü•ß This query shows the traffic distribution (High, Medium, Low) for phishing vs legitimate websites using pie charts.  


````python
plt.figure(figsize=(12,6))
plt.grid(True)
# Phishing sites
plt.subplot(1, 2, 1)
df[df["status"]=="phishing"]["traffic_category"].value_counts().plot.pie(
    autopct='%1.1f%%', colors=["salmon","orange","gold"])
plt.title("Phishing Sites Traffic Distribution")

# Legitimate sites
plt.subplot(1, 2, 2)
df[df["status"]=="legitimate"]["traffic_category"].value_counts().plot.pie(
    autopct='%1.1f%%', colors=["lightgreen","lime","green"])
plt.title("Legitimate Sites Traffic Distribution")

plt.tight_layout()
````

The plots compare phishing and legitimate sites. Phishing sites usually have longer, more complicated hostnames, while legitimate sites tend to keep them short. Also, legitimate sites mostly have higher page ranks, whereas phishing sites sit lower, showing they‚Äôre less trusted or recognized online.

### üìä This query plots the distribution of hostname lengths and page ranks for phishing vs legitimate websites using histograms.  


````python
plt.figure(figsize=(15, 6))
# 1
plt.subplot(1, 2, 1)
sns.histplot(x='length_hostname', hue='status', data=df, palette={'legitimate': 'lightgreen', 'phishing': 'lightpink'}, kde=True)
plt.title('Distribution of Hostname Lengths by Status')
plt.xlabel('Hostname Length')
plt.ylabel('Count')

# 2
plt.subplot(1, 2, 2)
sns.histplot(x='page_rank', hue='status', data=df, palette={'legitimate': 'lightgreen', 'phishing': 'lightpink'}, kde=True)
plt.title('Distribution of Page Ranks by Status')
plt.xlabel('Page Rank')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
````

This plot shows which features are most linked to phishing sites. Features in green tend to increase for phishing, while red ones are stronger in legitimate sites. It gives a quick sense of what characteristics make a site more likely to be phishing.

### üìä This query finds the top 10 features most correlated with phishing status and visualizes them using a bar plot.

````python
df['status_numeric'] = df['status'].map({'legitimate': 0, 'phishing': 1})

correlations = df.corr(numeric_only=True)['status_numeric'].sort_values(ascending=False)

top_correlations = correlations[1:11]

plt.figure(figsize=(12, 8))
colors = ['red' if c < 0 else 'green' for c in top_correlations]
sns.barplot(x=top_correlations.values, y=top_correlations.index, palette="pastel")

plt.title('Top 10 Features Correlated with Phishing Status', fontsize=16)
plt.xlabel('Correlation Coefficient', fontsize=12)
plt.ylabel('Features', fontsize=12)
plt.tight_layout()
plt.show()
````
